```stata
* for-loops going through carry-forward questions with string values to ensure that if respondent didn't indicate they offer a particular service/option, questions
* dependent on the answer are coded as "999" (code for valid missing), and "." if they are true missings

forvalues i = 1/26 {

replace E6_`i' = "." if (E5_`i' == 1 | E5_`i' == . | E5_999 == 1) & (E6_`i' == "999" | E6_`i' == "" | missing(E6_`i'))
replace E7_`i' = "." if (E5_`i' == 1 | E5_`i' == . | E5_999 == 1) & (E7_`i' == "999" | E7_`i' == "" | missing(E7_`i'))
replace E8_`i' = "." if (E5_`i' == 1 | E5_`i' == . | E5_999 == 1) & (E8_`i' == "999" | E8_`i' == "" | missing(E8_`i'))
replace E9_`i' = "." if (E5_`i' == 1 | E5_`i' == . | E5_999 == 1) & (E9_`i' == "999" | E9_`i' == "" | missing(E9_`i'))
replace E15_`i' = "." if (E14_`i' == 1 | E14_`i' == . | E5_999 == 1) & (E15_`i' == "999" | E15_`i' == "" | missing(E15_`i'))
replace E16_`i' = "." if (E14_`i' == 1 | E14_`i' == . | E5_999 == 1) & (E16_`i' == "999" | E16_`i' == "" | missing(E16_`i'))

replace C3_`i' = "." if (C2_`i' == 1 | C2_`i' == . | C2_999 == 1) & (C3_`i' == "999" | C3_`i' == "" | missing(C3_`i'))
replace C4_`i' = "." if (C2_`i' == 1 | C2_`i' == . | C2_999 == 1) & (C4_`i' == "999" | C4_`i' == "" | missing(C4_`i'))
replace C5_`i' = "." if (C2_`i' == 1 | C2_`i' == . | C2_999 == 1) & (C5_`i' == "999" | C5_`i' == "" | missing(C5_`i'))
replace C10_`i' = "." if (C9_`i' == 1 | C9_`i' == . | C2_999 == 1) & (C10_`i' == "999" | C10_`i' == "" | missing(C10_`i'))

replace NU2_`i' = 999 if (NextUp_CAFYES == 0 | NextUp_CAFYES == 999) & NU2_999 == 1

replace NU3_`i' = "." if (NU2_`i' == 1 | NU2_`i' == . | (NU2_999 == 1 & NextUp_CAFYES==1)) & (NU3_`i' == "999" | NU3_`i' == "" | missing(NU3_`i'))

replace NU4_`i' = "." if (NU2_`i' == 1 | NU2_`i' == . | (NU2_999 == 1 & NextUp_CAFYES==1)) & (NU4_`i' == "999" | NU4_`i' == "" | missing(NU4_`i'))
replace NU5_`i' = "." if (NU2_`i' == 1 | NU2_`i' == . | (NU2_999 == 1 & NextUp_CAFYES==1)) & (NU5_`i' == "999" | NU5_`i' == "" | missing(NU5_`i'))


replace NU8_`i' = . if (NU2_`i' == 1 | NU2_`i' == . | NU2_999 == 1) & (NU8_`i' == 999 | missing(NU8_`i'))
replace NU8_`i' = 999 if (NextUp_CAFYES == 0 | NextUp_CAFYES == 999) & NU8_999 == 1

replace NU9_`i' = "." if (NU2_`i' == 1 | NU2_`i' == . | NU2_999 == 1) & (NU9_`i' == "999" | NU9_`i' == "" | missing(NU9_`i'))
replace NU9_`i' = "999" if (NextUp_CAFYES == 0 | NextUp_CAFYES == 999)

replace CAL6_`i' = "." if (CAL5_`i' == 1 | CAL5_`i' == . | CAL5_999 == 1) & (CAL6_`i' == "999" | CAL6_`i' == "" | missing(CAL6_`i'))
replace CAL7_`i' = "." if (CAL5_`i' == 1 | CAL5_`i' == . | CAL5_999 == 1) & (CAL7_`i' == "999" | CAL7_`i' == "" | missing(CAL7_`i'))
replace CAL8_`i' = "." if (CAL5_`i' == 1 | CAL5_`i' == . | CAL5_999 == 1) & (CAL8_`i' == "999" | CAL8_`i' == "" | missing(CAL8_`i'))
replace CAL12_`i' = . if (CAL5_`i' == 1 | CAL5_`i' == . | CAL5_999 == 1) & (CAL12_`i' == 999 | missing(CAL12_`i'))
replace CAL13_`i' = "." if (CAL12_`i' == 1 | CAL12_`i' == . | CAL5_999 == 1) & (CAL13_`i' == "999" | CAL13_`i' == "" | missing(CAL13_`i'))

}

forvalues i = 1/15 {
	replace CAL3_`i' = . if CAL2_`i' == 1 & CAL3_`i' == 999	& CAL2_999 == 1
}
```

```stata
* cleaning unique identifying variables; adding district and region variables

gen COLLEGE_NAME_2 = upper(COLLEGE_NAME)
drop COLLEGE_NAME
rename COLLEGE_NAME_2 COLLEGE_NAME

merge m:1 COLLEGE_NAME using "Z:\projects\CCCCO\Data\Download2019\StataDatasets2019\colleges_district_names.dta", keepusing(COLLEGE_ID DISTRICT_ID DISTRICT_NAME)
drop if _merge==2
drop _merge

merge m:1 COLLEGE_NAME using "U:\brooks\CTE\Student Supports\Colleges By Region.dta"

keep if _merge==3

order DISTRICT_ID DISTRICT_NAME REGION COLLEGE_ID COLLEGE_NAME FIRST_NAME LAST_NAME COLLEGE_NAME EMAIL_ADDRESS EOPSCARE CalWORKs NextUp_CAFYES

save "Z:\projects\CCCCO\CTE\HE1803\Student Supports Survey Data.dta", replace
```

```stata
* creating subsets of master survey data file based on program (one file for respondents who reported for EOPS/CARE, one file for CalWORKs)

preserve

drop CAL*

keep if EOPSCARE == 1

save "Z:\projects\CCCCO\CTE\HE1803\Student Supports Survey Data EOPS.dta", replace

export excel using "U:\brooks\CTE\Student Supports\Student Supports Survey Data (EOPS).xlsx", firstrow(variables) replace

restore

preserve

keep DISTRICT_ID-INDEX CAL* ECAL* O*

keep if CalWORKs == 1

local CAL4 CAL4_1 CAL4_2 CAL4_3 CAL4_4 CAL4_5 CAL4_6 CAL4_7

foreach x of local CAL4 {
	replace `x' = "" if `x' == "Empty"
	destring `x', generate(`x'_2)
	drop `x'
	rename `x'_2 `x'
}

save "Z:\projects\CCCCO\CTE\HE1803\Student Supports Survey Data CALWORKS.dta", replace

export excel using "U:\brooks\CTE\Student Supports\Student Supports Survey Data (CALWORKS).xlsx", firstrow(variables) replace

restore
```
